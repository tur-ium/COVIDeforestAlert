```python
## Task 1. Check for correlations between deforestation cases and coronavirus cases

# Load libraries
import urllib2
import csv
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:76.0) Gecko/20100101 Firefox/76.0',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}

#Get data on numbers of COVID-19 cases
#Source European Centre for Disease Control: https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide
url = "https://opendata.ecdc.europa.eu/covid19/casedistribution/csv"

import pandas
covid_info=pandas.read_csv(url,parse_dates=[0],dayfirst=True) #Load and reverse index (so oldest to newest)
```

```python
covid_info = covid_info.sort_values("dateRep")
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

output_notebook() #Tell Bokeh to plot in Jupyter notebook


plot = figure(x_axis_type='datetime')

country_code = "BRA"
smoothed = covid_info[covid_info["countryterritoryCode"]==country_code].set_index("dateRep").cases.cumsum(skipna=True).rolling(2).mean()

source = ColumnDataSource(smoothed.to_frame())

#covid_info['cum_cases'] = covid_info[covid_info["countryterritoryCode"]=="BRA"].cases.cumsum(skipna=True)
#source = ColumnDataSource(covid_info)

hover = HoverTool()
hover.tooltips=[
    ('Date', '@dateRep{%F}'),
    ('Cases', '@cases'),
    ]
hover.formatters={'@dateRep': 'datetime',"@cases":"numeral"}
plot.add_tools(hover)
plot.yaxis.axis_label="Total number of cases"
plot.xaxis.axis_label="Date"
plot.title.text=country_code
plot.line(x="dateRep", y="cases",source=source,color="black")
show(plot)
```
